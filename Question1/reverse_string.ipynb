```python
def reversed_str(string: str) -> str:
    """
    This is a function, so reversed the sting.
    Args:
        Input_string: Only take the string from the user.
    Returns:
        String: Return reversed version of given string.
    """
    
    if not isinstance(string,str):
        raise TypeError("Type Error: Ginven input is not a string")

    if string.isdigit():
        raise ValueError("Value Error: Numbers are not allowed, Please enter a string.")

    reversed_str=""
    for i in range(len(string)):
        reversed_str = string[i]+reversed_str
    print(f"Your reversed string: {reversed_str}")
    return reversed_str

if __name__=="__main__":
    try:
        
        #taking input from the user
        user=input("Enter a string to reverse: ")

        #calling the function 
        reversed_str(user)
        
    except (TypeError,ValueError as e:
        print(e)
    except Exception as e :
         print(e)
```
