```python
def large_num(list1 : list) -> list:
    """
    Function is return the largest number in the present list if the given numbers in a list are not equal.
    In case all numbers in the list are same this function is return none.
    Args:
        Input(lis1: int,float): This fuction is accept only int or float type.
    Returns:
        Largest number or None: In case numbers in list are not equal returns Largest number
                                other then return None
    """
    
    # Validate input not list
    if not isinstance(list1,list):
        raise TypeError("TypeError: Function take only list.")
        
    # Validate input must be one arguments
    if not list1:
        raise ValueError("ValueError: Function can must one agruments not empty")
        
    # Validate numbers in input list must be int or float
    for item in list1:
        if not isinstance(item,(int,float)):
            raise TypeError("TypeError: Input are not int or float.")

    

    # LOgic bulding for find largest number in list
    
    largest_num = list1[0]
    all_same=True
    for item in list1:
        if item != largest_num:
            all_same = False
        if item > largest_num : 
            largest_num = item

    
    # return None if all numbers are name in list
    if all_same:
        return None
    # return largest number if all numbers are not equal
    else :
        return largest_num

if __name__=="__main__":
    try:
        # Take input from the user
        user_input=(input("Enter numbers in a list seperated by space which is int or float : "))
        user_input = [float(i) for i in user_input.split()]

        
        # Calling the function
        result=large_num(user_input)

        print(f"The largest number in list {user_input} is : {result}")

    except (TypeError,ValueError) as error:
        print(error)
    except Exception as error:
        print(error)
```
