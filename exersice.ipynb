```python
import pyttsx3
from datetime import datetime
engine = pyttsx3.init('sapi5')   
engine.setProperty('rate', 125)


def speak(text):
    engine.say(text)


def time_based_greeting():
    hour = datetime.now().hour

    if 5 <= hour < 12:
        speak("Good morning.")
        print("Good morning!")
    elif 12 <= hour < 17:
        speak("Good afternoon.")
        print("Good afternoon!")
    elif 17 <= hour < 21:
        speak("Good evening.")
        print("Good evening!")
    else:
        speak("Good night.")
        print("Good night!")
        
time_based_greeting()

while True:
    speak("Enter the product name : ")
    engine.runAndWait()
    user = input("Enter the product name : ")
    
    
    if user in product_price:
        speak(f"{user} price : {product_price[user]} INR")
        print(f"{user} price : {product_price[user]} INR")
        speak("Do you want to update product price : ")
        engine.runAndWait()
        yes_form_user=input("Do you want to update product price : ")
    
        if yes_form_user== "yes" or yes_form_user=="Yes" or yes_form_user=="YES" :
            speak("Enter update price : ")
            # engine.runAndWait()
            update = int(input("Enter update price : "))
            engine.runAndWait()
            product_price[user] = update
            speak(f"Update price of {user} : {update} INR")
            print(f"Update price of {user} : {update} INR")
            
        elif yes_form_user=="no" or yes_form_user=="No" or yes_form_user=="NO":
            speak("Thank you for using me.")
            print("Thank you for using me.")
            
        else:
            speak("You are enter Invalid input.\nPlease enter Yes or No.")
            print("You are enter Invalid input.\nPlease enter Yes or No.")
            
    else :
        speak(f"{user} is not found on system")
        print(f"{user} is not found on system")
        speak(f"Do you want to update {user} product on system : ")
        engine.runAndWait()
        update_product=input(f"Do you want to update {user} product on system : ")
        if update_product== "yes" or update_product== "Yes" or update_product== "YES" :
            speak(f"Enter the price of {user} : ")
            
            price=int(input(f"Enter the price of {user} : "))
            engine.runAndWait()
            product_price.update([(user,price)])
            # product_price[user]=price
            speak(f"Update product and price is : {user} : {price} INR")
            print(f"Update product and price is : {user} : {price} INR")
            
            # print(f"Update price : {update} INR")
        elif update_product=="no" or update_product =="No" or update_product=="NO":
            speak("Thank you for using me.")
            print("Thank you for using me.")
            
        else : 
            speak("You are enter wrong input.\nPlease enter only yes or no.")
            print("You are enter wrong input.\nPlease enter only yes or no.")

    
    while True:
        speak("Do you want to check another product. Enter only yes or not.")
        engine.runAndWait()
        again = input("Do you want to check another product (yes/no): ").strip().lower()
    
        if again == "yes":
            speak("Okay.")
            break          # exit inner loop â†’ continue main program
    
        elif again == "no":
            speak("Okay, exiting the program. Goodbye.")
            print("Program exited.")
            
            exit()
    
        else:
            speak("Invalid input. Please enter yes or no.")
            print("Invalid input. Please enter yes or no.")
            
            
```
