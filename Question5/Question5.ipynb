```python
def factorial(num: int) -> int :
    """
    This function can find the factorial of given number.
    Args:
        This function take integer number
    Returns:
        This functioin is return the factorial of given number.
    """

    # Validate input is integer
    if not isinstance(num, int):
        raise TypeError("TypeError: input only integer.")

    if not num:
        raise ValueError("ValueError: input must be take one argument.")

    
    fact = 1

    for i in range(1,num+1):
        fact = fact*i

    # Return factorial of num
    return fact

# main exicution
if __name__ == "__main__":
    try:
        # take input from user
        user_input=int(input("Enter the number whose you want to factorial: "))
        # calling funtion 
        result = factorial(user_input)
        # print factorial
        print(f"Factorial of {user_input} is {result}")

    except (TypeError,ValueError) as error:
        print(error)

    except Exception as error:
        print(error)
```
