```python
num=int(input("Enter a number : "))
if num > 1:
    for i in range(2,int(num**0.5)+1):
        if num % i == 0:
            print(f"{num} is NOT a prime number.")
            break
    else:
        print(f"{num} is a prime number.")

else:
    print(f"{num} is NOT a prime number.")
```

```python
def range_prime(start: int,stop: int) -> int:
    """
    This function is print all prime numbers between given limit
    Args:
        This function is take limit start or stop 
    Returns:
        True given number is prime 
    """
    if not isinstance(start,int):
        raise TypeError("take only integer.")

    for num in range(start,stop):
        is_prime=True
        for i in range(2,int(num**0.5)+1):
            if num % i == 0:
                is_prime = False
        if is_prime:
            print(num)

if __name__ == "__main__" :
    
    range_prime(100,250)
```
