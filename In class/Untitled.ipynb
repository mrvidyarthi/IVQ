```python
def rev_str(string: str):
    """
    Args: 
        Input String: Take only a string.
    Returns:
        String: Returns the reveres version of given string.
    """
    if not isinstance(string,str):
        raise TypeError("Type Error: Given input is not a string.")
    if string.isdigit():
        raise ValueError("Value Error: Numbers are not allowed.")
    rev_str = ""
    for char in string:
        rev_str = char + rev_str
    print(rev_str)

    return rev_str
    
if __name__=="__main__":

    try:
        # Taking input from the user
        user = input("Enter a string: ")
        
        # Calling the function
        rev_str(user)
        
    except TypeError as e:
        print(e)
    except ValueError as e :
        print(e)
    except Exception as e :
        print(e)
    finally:
        print("Thankyou for using me")
        
```
